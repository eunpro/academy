기존의 프로그래밍 방식

```python
def celsius_to_fahrenheit(x):
  return x * 1.8 + 32
```

```python
celsius_value = int(input('섭씨온도를 입력하세요.'))
print('화씨온도로', celsius_to_fahrenheit(celsius_value))
```

머신러닝 방식

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_C = np.array(range(0,100)) #섭씨
data_F = celsius_to_fahereheit(data_C) #화씨
print(data_C)
print(data_F)
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(1,)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop')
print(model.summary())
```

```python
scaled_data_C =data_C/100
scaled_data_F =data_F/100
print(scaled_data_C)
print(scaled_data_F)
```

```python
model.predict([0.01])
```

```python
model.save('before_learning.h5') #1차 모델 저장
```

```python
#데이터 학습시키기
fit_hist = model.fit(scaled_data_C, scaled_data_F, epochs=1000) #입력데이터 1000개를 만들어라
```

```python
print(model.predict([0.01]))
```

```python
model.save('after_learning.h5')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show() #colab에서는 안해도 되지만 다른 프로그램에서는 실행해야함
```

####잡음이 있는 실제 데이터

```python
noise = np.array(np.random.normal(0,0.05,100))
#normal: 정규분포따름, 평균 0이고 표준편차가 0.05인 잡음 100개 생성
#정규분포에서 ±3σ 내에 99% 데이터 존재
noise
```

```python
noised_scaled_data_F = np.array([])
for data in scaled_data_F:
  noised_scaled_data_F = np.append(
      noised_scaled_data_F, np.random.normal(
          0,0.05,100) + data )
  
print(noised_scaled_data_F)  
print(len(noised_scaled_data_F)) #100개씩 100개의 data
```

```python
noised_scaled_data_C = []
for data in range(0,100):
  for i in range(0,100):
    noised_scaled_data_C.append(data)
noised_scaled_data_C = np.array(noised_scaled_data_C)
noised_scaled_data_C = noised_scaled_data_C/100
print(noised_scaled_data_C)
print(len(noised_scaled_data_C))    
```

```python
plt.scatter(x=noised_scaled_data_C, y= noised_scaled_data_F)  #산점도: X,y값으로 점찍는 것
plt.show()
```

```python
fig = plt.figure(figsize=(50,50)) #그래프 사이즈
ax = fig.add_subplot(111)
ax.scatter(x=noised_scaled_data_C, 
           y=noised_scaled_data_F,
           alpha=0.2, s=200, marker='+') #alpha투명도,marker표시형태 
plt.show() #정규분포를 하고 있는 (x값 100개) * (y값 100개) 생성됨
```

```python
model2 = Sequential() #keras의 순차적 모델
model2.add(InputLayer(input_shape=(1,)))
model2.add(Dense(1)) #모델 뇌세포 1개 짜리
model2.compile(loss='mse', optimizer='rmsprop')
print(model2.summary())
#결과 param => bias, weight
```

```python
print(model2.predict([0.01]))
```

```python
fit_hist = model2.fit(noised_scaled_data_C,
                      noised_scaled_data_F, epochs=20)#fit(입력, 출력, 학습횟수)
```

```python
model2.save('noised_after_learning.h5')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```
