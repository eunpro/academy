```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
#훈련데이터
training_data = np.array(
    [[0,0], [0,1], [1,0], [1,1]], 'float32')
#정답데이터
target_data = np.array(
    [[0],[0],[0],[1]],'float32')
```

```python
model = Sequential()
model.add(Dense(32,input_dim=2,
                activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='mse', optimizer='adam',
              metrics = ['binary_accuracy'])
print(model.summary())
```

image.png

1겹 : input 2개 * binary 32개 + binary 32개 => 96개

2겹 : input 32개 + binary 1개 => 33개

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1) #데이터 적어서 학습 많이
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
#and게이트 입력값 00 => 0 / 10 =>0 / 01=>0 / 11=>1
inp = list(map(int, input().split()))
qwe = np.array(inp)
print('입력 값 :', qwe)
qwe = qwe.reshape(1,2)
print('reshape :', qwe)
print('결과 값 :', model.predict(qwe)[0][0].round())
```
